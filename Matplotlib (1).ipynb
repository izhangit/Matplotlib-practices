# Matplotlib

# Line Plot


```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color = 'yellow', linestyle = ':', linewidth = '7')
plt.title('Line Plot')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1, color = 'yellow', linestyle = ':', linewidth = '7')
plt.plot(x,y2, color = 'red', linestyle = '-.', linewidth = '8')
plt.title(' Line Plot')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1, color = 'red', linestyle = ':',  linewidth = 4)

plt.subplot(1,2,2)
plt.plot(x,y2, color = 'g', linewidth = 4)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1, color = 'blue', linestyle = ':',  linewidth = 4)

plt.subplot(2,1,2)
plt.plot(x,y2, color = 'orange', linewidth = 4)
# plt.grid(True)

plt.show()
```

# Bar plot

```python
students = {"Bob" : 45 ,"Sam" : 77, "Matt" : 23}

names = list(students.keys())
values = list(students.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title('Vertical Bar Plot')
plt.xlabel('Names')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values, color = 'red')
plt.title('Horizontal Bar Plot')
plt.xlabel('Marks')
plt.ylabel('Names')
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
# creating a basic Scatter-plot
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,3,1,5,7,8,9,4,6]
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
```

```python
plt.scatter(x,y, marker = '*' , color = 'red', s = 150)
plt.title("Scatter Plot")
plt.show()
```

```python
# Adding two makers in the same Plot
```

```python
z = [1,2,3,6,5,8,4,7,9]

plt.scatter(x,y, marker = '*', color = 'blue' , s = 99)
plt.scatter(x,z, marker = '.', color = 'red' , s = 200)

plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y, marker = '*', color = 'blue', s = 45)
plt.subplot(1,2,2)
plt.scatter(x,z, marker = '.', color = 'black' , s = 55)

plt.show()
```

# Pie-Chart

```python
fruits = ['Apple' , ' Orange' , ' Banana ' , ' Mango ']
quantity = [67,34,89,29]

```

```python
plt.pie(quantity,labels = fruits)
plt.show()
```

```python
plt.pie(quantity,labels = fruits, autopct = '%0 1f%%')
plt.show()
```
